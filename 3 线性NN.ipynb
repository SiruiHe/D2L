# 线性NN

## 知识点
### 线性回归
- 这是一个仿射变换：线性变换 + 偏置量（平移）
$$\mathrm{price} = w_{\mathrm{area}} \cdot \mathrm{area} + w_{\mathrm{age}} \cdot \mathrm{age} + b.$$
- 模型参数：$\mathbf{w}$、$b$
- 损失函数：MSE
- 存在解析解：最小二乘
- 优化方法：梯度下降、小批量梯度下降等
- 超参数：批量大小、学习率等
- 泛化（generalization）
- 假设：随机噪声，正态分布
- 神经网络的层（layer）
  - 全连接层（fully-connected）或稠密层（dense）
- PyTorch 高级API
  - `data`提供数据处理、`nn`提供神经网络层&常见函数
  - 通过`_`结尾的方法原地修改，可用于初始化参数
- 分类问题
  - one-hot编码：一组单位正交向量
  - 校准（calibration）：采用softmax函数，非负数、和为1
  - 过softmax只会改变大小不会改变次序，所以仍然算线性模型
  - 分类问题的损失函数：交叉熵（cross-entropy loss），基于最大似然估计

```python
%matplotlib inline
import math
import time
import numpy as np # 避免使用for，而使用线性代数来处理
import torch
from d2l import torch as d2l

n = 10000
a = torch.ones([n])
b = a.clone()

class Timer:  #@save
    """记录多次运行时间"""
    def __init__(self):
        self.times = []
        self.start()

    def start(self):
        """启动计时器"""
        self.tik = time.time()

    def stop(self):
        """停止计时器并将时间记录在列表中"""
        self.times.append(time.time() - self.tik)
        return self.times[-1]

    def avg(self):
        """返回平均时间"""
        return sum(self.times) / len(self.times)

    def sum(self):
        """返回时间总和"""
        return sum(self.times)

    def cumsum(self):
        """返回累计时间"""
        return np.array(self.times).cumsum().tolist()
    
c = torch.zeros(n)
timer = Timer()
for i in range(n):
    c[i] = a[i] + b[i]
f'{timer.stop():.5f} sec'
```

```python
# 正态分布
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

x = np.arange(-7, 7, 0.01) # 自变量范围 -7 ~ 7
params = [(0, 1), (0, 2), (3, 1)]
d2l.plot(x, [normal(x, mu, sigma) for mu, sigma in params], xlabel='x',
         ylabel='p(x)', figsize=(4.5, 2.5),
         legend=[f'mean {mu}, std {sigma}' for mu, sigma in params])
```

```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l

def synthetic_data(w, b, num_examples): #@save
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape(-1, 1)

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
print('features:', features[0], '\nlabel:', labels[0])
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, (1)].detach().numpy(), labels.detach().numpy(), 1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)]
        )
        # 返回一个迭代器:每次迭代时，yield会返回一个值，并在函数中暂停执行，直到下一次迭代请求值时再继续执行。
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break # 仅处理第一个X，y数据就退出循环
```

```python
w = torch.normal(0, 0.01, size=(2,1),requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# 线性网络模型
def linreg(X, w, b): #@save
    return torch.matmul(X, w) + b
# y = Wx + b

# 损失函数
def squared_loss(y_hat, y): #@save
    # 以batch方式处理数据时，输出shape为（size, dim)
    # 但是真实标签y是可能只是一维数组，因此需要用reshape来使其与y_hat匹配
    return(y_hat - y.reshape(y_hat.shape)) ** 2 / 2

# 优化算法：小批量随机梯度下降
def sgd(params, lr, batch_size):
    with torch.no_grad():
        for param in params:
            # 梯度「下降」
            param -= lr * param.grad / batch_size
            param.grad.zero_() # 原地将参数的梯度清零，否则会累积
```

**执行以下循环**：
* 初始化参数
* 重复以下训练，直到完成
    * 计算梯度$\mathbf{g} \leftarrow \partial_{(\mathbf{w},b)} \frac{1}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} l(\mathbf{x}^{(i)}, y^{(i)}, \mathbf{w}, b)$
    * 更新参数$(\mathbf{w}, b) \leftarrow (\mathbf{w}, b) - \eta \mathbf{g}$

```python
# 设置超参数
lr = 0.03
num_epochs = 3
net = linreg # 网络模型
loss = squared_loss # 损失函数

# 每个epoch的训练过程
for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y) # X和y的小批量损失
        # y的shape是（batch_size, 1)，而非标量
        # 但是.backward()方法需要一个标量作为反向传播的起点
        # 使用sum将批量的loss向量合并为一个标量，从而进行参数更新
        l.sum().backward() # 优化目标为这一小批量的loss之和最小
        # SGD更新模型参数
        sgd([w, b], lr, batch_size)
    with torch.no_grad(): # 禁止梯度计算
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}， loss {float(train_l.mean()):f}')
```

```python
# 利用DL库的简洁实现
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)

'''pytorch数据迭代器'''
def load_array(data_arrays, batch_size, is_train=True):
    dataset = data.TensorDataset(*data_arrays) # *前缀用于解包参数列表，因此可以灵活传递任意数量的参数
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 10
# DataLoader类型
data_iter = load_array((features, labels), batch_size)

# 用iter构造迭代器，用next获取下一项
next(iter(data_iter))
```

```python
from torch import nn
net = nn.Sequential(nn.Linear(2, 1)) # input, output shape
```

```python
net[0].weight.data.normal_(0, 0.01) # weight初始化
net[0].bias.data.fill_(0) # bias初始化
loss = nn.MSELoss() # 均方误差
trainer = torch.optim.SGD(net.parameters(), lr=0.03)

num_epochs = 3
for epoch in range(num_epochs):
    # 一个batch（小批量更新）
    for X, y in data_iter:
        l = loss(net(X), y) # MSELoss
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features), labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```

```python
# 从0开始实现
import torch
from IPython import display # 交互式解释器
from d2l import torch as d2l

batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(
    batch_size=batch_size)
```

```python
num_inputs = 784
num_outputs = 10

W = torch.normal(0, 0.01, size=(num_inputs, num_outputs), requires_grad=True)
b = torch.zeros(num_outputs, requires_grad=True)
```

```python
def softmax(X):
    X_exp = torch.exp(X)
    partition = X_exp.sum(1, keepdim=True) # 保留为1的维度
    return X_exp / partition
```

```python
X = torch.normal(0, 1, (2, 5))
X_prob = softmax(X)
X_prob, X_prob.sum(1) # 每行是一个样本，每行的和为1（即sum(1)=1）
```

```python
def net(X):
    return softmax(torch.matmul(X.reshape((-1, W.shape[0])), W) + b)
```

```python
# 分类问题定义交叉熵函数
y = torch.tensor([0, 2]) # 正确标签
y_hat = torch.tensor([[0.1, 0.3, 0.6], [0.3, 0.2, 0.5]]) # 假想的预测结果
# 高级索引方法，[0, 1]行索引，y为列索引；由交叉点上的元组构建新张量返回
y_hat[[0, 1], y] # 第一个样本中预测正确的概率，第二个样本中预测正确的概率
```

```python
def cross_entropy(y_hat, y):
    # range分别选择第1、2、3...行对应的y_hat
    return -torch.log(y_hat[range(len(y_hat)), y])

cross_entropy(y_hat, y)
```

```python
# 计算精度（正确预测的数量）
def accuracy(y_hat, y):
    if len(y_hat.shape) > 1 and y_hat.shape[1] > 1:
        y_hat = y_hat.argmax(axis=1) # 去每行中最大元素的索引
        cmp = y_hat.type(y.dtype) == y # 用==判断是否正确
        return float(cmp.type(y.dtype).sum())
```

```python
class Accumulator:  #@save
    """在n个变量上累加"""
    def __init__(self, n):
        self.data = [0.0] * n

    def add(self, *args):
        self.data = [a + float(b) for a, b in zip(self.data, args)]

    def reset(self):
        self.data = [0.0] * len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]
```

```python
def evaluate_accuracy(net, data_iter):
    '''在指定数据集上的精度'''
    if isinstance(net, torch.nn.Module):
        net.eval()
    metric = Accumulator(2) # 对2个变量进行累加：正确预测数、预测总数
    with torch.no_grad():
        for X, y in data_iter:
            metric.add(accuracy(net(X), y), y.numel()) # numel()返回张量中元素的总数
    return metric[0] / metric[1]
```

```python
evaluate_accuracy(net, test_iter)
```

```python
def train_epoch_ch3(net, train_iter, loss, updater): #@save
    if isinstance(net, torch.nn.Module):
        net.train()
    metric = Accumulator(3)
    for X, y in train_iter:
        y_hat = net(X)
        l = loss(y_hat, y)
        if isinstance(updater, torch.optim.Optimizer):
            updater.zero_grad() # 梯度清零
            l.mean().backward() # 使用了mean()，所以与批量大小无关
            updater.step() # 是torch内置的Optimizer对象，调用`step`使梯度更新
        else:
            l.sum().backward()
            updater(X.shape[0]) # 自定义的updater负责更新的全部逻辑，传入大小作为参数
        metric.add(float(l.sum()), accuracy(y_hat, y), y.numel())

    return metric[0] / metric[2], metric[1] / metric[2]
```

```python
class Animator:  #@save
    """在动画中绘制数据"""
    def __init__(self, xlabel=None, ylabel=None, legend=None, xlim=None,
                 ylim=None, xscale='linear', yscale='linear',
                 fmts=('-', 'm--', 'g-.', 'r:'), nrows=1, ncols=1,
                 figsize=(3.5, 2.5)):
        # 增量地绘制多条线
        if legend is None:
            legend = []
        d2l.use_svg_display()
        self.fig, self.axes = d2l.plt.subplots(nrows, ncols, figsize=figsize)
        if nrows * ncols == 1:
            self.axes = [self.axes, ]
        # 使用lambda函数捕获参数
        self.config_axes = lambda: d2l.set_axes(
            self.axes[0], xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
        self.X, self.Y, self.fmts = None, None, fmts

    def add(self, x, y):
        # 向图表中添加多个数据点
        if not hasattr(y, "__len__"):
            y = [y]
        n = len(y)
        if not hasattr(x, "__len__"):
            x = [x] * n
        if not self.X:
            self.X = [[] for _ in range(n)]
        if not self.Y:
            self.Y = [[] for _ in range(n)]
        for i, (a, b) in enumerate(zip(x, y)):
            if a is not None and b is not None:
                self.X[i].append(a)
                self.Y[i].append(b)
        self.axes[0].cla()
        for x, y, fmt in zip(self.X, self.Y, self.fmts):
            self.axes[0].plot(x, y, fmt)
        self.config_axes()
        display.display(self.fig)
        display.clear_output(wait=True)
```

```python
def train_ch3(net, train_iter, test_iter, loss, num_epochs, updater): #@save
    animator = Animator(xlabel='epoch', xlim=[1, num_epochs], ylim=[0.3, 0.9],
                        legend=['train loss', 'train acc', 'test acc'])
    for epoch in range(num_epochs):
        train_metrics = train_epoch_ch3(net, train_iter, loss,updater)
        test_acc = evaluate_accuracy(net, test_iter)
        animator.add(epoch + 1, train_metrics + (test_acc,))
    train_loss, train_acc = train_metrics
    assert train_loss < 0.5, train_loss
    assert train_acc <= 1 and train_acc > 0.7, train_acc
    assert test_acc <= 1 and test_acc > 0.7, test_acc
```

```python
lr = 0.1
num_epochs = 10

def updater(batch_size):
    return d2l.sgd([W, b], lr, batch_size)
```

```python
train_ch3(net, train_iter, test_iter, cross_entropy, num_epochs, updater)
```

```python
def predict_ch3(net, test_iter, n=6): #@save
    for X, y in test_iter:
        break
    trues = d2l.get_fashion_mnist_labels(y)
    preds = d2l.get_fashion_mnist_labels(net(X).argmax(axis=1))
    titles = [true + '\n' + pred for true, pred in zip(trues, preds)]
    d2l.show_images(
        X[0:n].reshape((n, 28, 28)),1, n, titles=titles[0:n]
    )

predict_ch3(net, test_iter)
```
